```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Machine Downtime.csv')
df.head()
```

```python
df.columns
```

```python
#Removing unnecessary columns
df.drop(columns=['Date','Machine_ID','Assembly_Line_No'],axis=1,inplace=True)
df.head(2)
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.nunique() #no of unique values in each column
```

```python
df.isnull().sum() #total null values in each feature
```

```python
df.dropna(inplace=True) #removing rows with null values
df
```

```python
df.shape
```

```python
df.notnull().sum() # 0 missing values
```

```python
#Coverting categorical data into numerical data
df.loc[df['Downtime']=='Machine_Failure','Downtime']=1
df.loc[df['Downtime']=='No_Machine_Failure','Downtime']=0
df['Downtime'] = pd.to_numeric(df['Downtime'])
df.head(2)
```

```python
X=df.drop('Downtime',axis=1)
y=df['Downtime']
```

```python
df['Downtime'].value_counts() #data is balanced
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape
```

```python
y_train
```

```python
#Feature Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Model training
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression()
model=regressor.fit(X_train,y_train)
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python
metrics=confusion_matrix(y_test,y_pred)
print(metrics)
print(classification_report(y_test,y_pred))
```

```python
#Hyperparameter Tuning using GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
params=[{'C':[1,5,10,15,20,30,40,50]},{'max_iter':[100,150,200,300]}]
```

```python
model1=GridSearchCV(model,param_grid=params,scoring='f1',cv=5)
model1.fit(X_train,y_train)
```

```python
model1.best_params_
```

```python
y_pred=model1.predict(X_test)
```

```python
#Final accuracy
metrics=confusion_matrix(y_test,y_pred)
print(metrics)
print(classification_report(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
def classification_scores(org,pred):
    print("accuracy score: ",accuracy_score(org,pred))
    print("precision score: ",precision_score(org,pred))
    print("recall score: ",recall_score(org,pred))
    print("f1 score: ",f1_score(org,pred))
```

```python
classification_scores(y_test,y_pred)
```
